```python
import numpy as np
import skfuzzy as fuzz
import matplotlib .pyplot as plt
import sys
```

```python
x_temp = np.arange(0, 50, 0.5)
x_soil_moist = np.arange(0, 100, 1)
x_humidity = np.arange(0, 100, 1)
x_light_intensity = np.arange(0, 1000, 10)
x_irrigation = np.arange(0, 60, 1)

print("x_temp: " + str(x_temp))
print("x_soil_moist: " + str(x_soil_moist))
print("x_humidity"+ str(x_humidity))
print("x_light_intensity"+ str(x_light_intensity))
print("x_irrigation"+ str(x_irrigation))
```

# Linguistic Variable

```python
# Temperature Membership Functions
temp_cold = fuzz.trapmf(x_temp,   [0, 0, 5 ,12])
print("temp_cold: " + str(temp_cold))
temp_fresh = fuzz.trimf(x_temp,   [7, 15, 22])
print("temp_fresh: " + str(temp_fresh))
temp_normal = fuzz.trimf(x_temp,   [16, 23, 30])
print("temp_normal: " + str(temp_normal))
temp_hot = fuzz.trimf(x_temp,   [25 ,31, 38])
print("temp_hot: " + str(temp_hot))
temp_vHot = fuzz.trapmf(x_temp,  [33 ,41 ,50 ,50])
print("temp_vHot: " + str(temp_vHot))
```

```python
# Soil Moisture Membership Functions
soil_moist_Dry = fuzz.trapmf(x_soil_moist,  [0 ,0, 20, 40])
print("soil_moist_Dry: " + str(soil_moist_Dry))
soil_moist_humide = fuzz.trimf(x_soil_moist,   [25 , 45 , 65])
print("soil_moist_humide: " + str(soil_moist_humide))
soil_moist_wet = fuzz.trapmf(x_soil_moist,  [50 ,75, 100, 100])
print("soil_moist_wet: " + str(soil_moist_wet))
```

```python
# Humidity Membership Functions
humidity_low = fuzz.trapmf(x_humidity, [0, 0, 35, 50])
print("humidity_low: " + str(humidity_low))
humidity_medium = fuzz.trimf(x_humidity, [35, 50, 70])
print("humidity_medium: " + str(humidity_medium))
humidity_high = fuzz.trapmf(x_humidity, [52.5, 70, 100, 100])
print("humidity_high: " + str(humidity_high))
```

```python
# Light Intensity Membership Functions
light_very_dark = fuzz.trapmf(x_light_intensity, [0, 0, 50, 100])
print("light_very_dark: " + str(light_very_dark))
light_dark = fuzz.trapmf(x_light_intensity, [50, 100, 300, 350])
print("light_dark: " + str(light_dark))
light_normal = fuzz.trapmf(x_light_intensity, [300, 400, 600, 700])
print("light_normal: " + str(light_normal))
light_bright = fuzz.trapmf(x_light_intensity, [650, 700, 900, 950])
print("light_bright: " + str(light_bright))
light_very_bright = fuzz.trapmf(x_light_intensity, [900, 950, 1000, 1000])
print("light_very_bright: " + str(light_very_bright))
```

```python
irrigation_short = fuzz.trapmf(x_irrigation,   [0 , 0 ,10 ,20])
print("irrigation_short: " + str(irrigation_short))
irrigation_medium = fuzz.trapmf(x_irrigation,  [13, 20, 35, 42])
print("irrigation_medium: " + str(irrigation_medium))
irrigation_long = fuzz.trapmf(x_irrigation,   [35, 45 ,60, 60])
print("irrigation_long: " + str(irrigation_long))
```

# Irrigation Fuzzy System Class

```python
class Fuzzy_Irrigation():

    def __init__(self,CURRENT_TEMP,CURRENT_SOIL_MOIST,CURRENT_HUMIDITY,CURRENT_LIGHT_INTENSITY):
        self.CURRENT_TEMP = CURRENT_TEMP
        self.CURRENT_SOIL_MOIST = CURRENT_SOIL_MOIST
        self.CURRENT_HUMIDITY = CURRENT_HUMIDITY
        self.CURRENT_LIGHT_INTENSITY = CURRENT_LIGHT_INTENSITY

    def fuzzification(self):
        self.tempCold = fuzz.interp_membership(x_temp, temp_cold, self.CURRENT_TEMP)
        print("tempCold: " + str(self.tempCold))
        self.tempFresh = fuzz.interp_membership(x_temp, temp_fresh, self.CURRENT_TEMP)
        print("tempFresh: " + str(self.tempFresh))
        self.tempNormal = fuzz.interp_membership(x_temp, temp_normal, self.CURRENT_TEMP)
        print("tempNormal: " + str(self.tempNormal))
        self.tempHot = fuzz.interp_membership(x_temp, temp_hot, self.CURRENT_TEMP)
        print("tempHot: " + str(self.tempHot))
        self.tempVeryHot = fuzz.interp_membership(x_temp, temp_vHot, self.CURRENT_TEMP)
        print("tempVeryHot: " + str(self.tempVeryHot)+"\n")

        self.soilMoistDry = fuzz.interp_membership(x_soil_moist, soil_moist_Dry, self.CURRENT_SOIL_MOIST)
        print("soilMoistDry: " + str(self.soilMoistDry))
        self.soilMoistHumide = fuzz.interp_membership(x_soil_moist, soil_moist_humide, self.CURRENT_SOIL_MOIST)
        print("soilMoistHumide: " + str(self.soilMoistHumide))
        self.soilMoistWet = fuzz.interp_membership(x_soil_moist, soil_moist_wet, self.CURRENT_SOIL_MOIST)
        print("soilMoistWet: " + str(self.soilMoistWet)+"\n")

        self.humidityLow = fuzz.interp_membership(x_humidity, humidity_low, self.CURRENT_HUMIDITY)
        print("humidityLow: " + str(self.humidityLow))
        self.humidityMedium = fuzz.interp_membership(x_humidity, humidity_medium, self.CURRENT_HUMIDITY)
        print("humidityMedium: " + str(self.humidityMedium))
        self.humidityHigh = fuzz.interp_membership(x_humidity, humidity_high, self.CURRENT_HUMIDITY)
        print("humidityHigh: " + str(self.humidityHigh)+"\n")

        self.lightintensityVeryDark = fuzz.interp_membership(x_light_intensity, light_very_dark, self.CURRENT_LIGHT_INTENSITY)
        print("lightintensityVeryDark: " + str(self.lightintensityVeryDark))
        self.lightintensityDark = fuzz.interp_membership(x_light_intensity, light_dark, self.CURRENT_LIGHT_INTENSITY)
        print("lightintensityDark: " + str(self.lightintensityDark))
        self.lightintensityNormal = fuzz.interp_membership(x_light_intensity, light_normal, self.CURRENT_LIGHT_INTENSITY)
        print("lightintensityNormal: " + str(self.lightintensityNormal))
        self.lightintensityBright = fuzz.interp_membership(x_light_intensity, light_bright, self.CURRENT_LIGHT_INTENSITY)
        print("lightintensityBright: " + str(self.lightintensityBright))
        self.lightintensityVeryBright = fuzz.interp_membership(x_light_intensity, light_very_bright, self.CURRENT_LIGHT_INTENSITY)
        print("lightintensityVeryBright: " + str(self.lightintensityVeryBright))

    def inference(self):
        #Fuzzy Inference for Short Irrigation
        self.Rule_1 = np.fmin(np.fmin(self.tempCold,self.soilMoistDry),irrigation_short)
        self.Rule_2 = np.fmin(np.fmin(self.tempCold,self.soilMoistHumide),irrigation_short)
        self.Rule_3 = np.fmin(np.fmin(self.tempCold,self.soilMoistWet),irrigation_short)
        self.Rule_4 = np.fmin(np.fmin(self.tempFresh,self.soilMoistHumide),irrigation_short)
        self.Rule_5 = np.fmin(np.fmin(self.tempFresh,self.soilMoistWet),irrigation_short)
        self.Rule_6 = np.fmin(np.fmin(self.tempNormal,self.soilMoistWet),irrigation_short)
        self.Rule_7 = np.fmin(np.fmin(self.humidityHigh,self.lightintensityDark),irrigation_short)
        self.Rule_8 = np.fmin(np.fmin(self.humidityHigh,self.lightintensityVeryDark),irrigation_short)
        self.Rule_9 = np.fmin(np.fmin(np.fmin(self.tempFresh,self.lightintensityVeryDark),self.soilMoistHumide),irrigation_short) 
        self.Rule_10 = np.fmin(np.fmin(np.fmin(self.tempCold,self.humidityHigh),self.soilMoistWet),irrigation_short) 
        self.Rule_11 = np.fmin(np.fmin(np.fmin(np.fmin(self.tempCold,self.soilMoistDry),self.humidityHigh),self.lightintensityVeryDark),irrigation_short)
        self.Rule_12 = np.fmin(np.fmin(np.fmin(np.fmin(self.tempNormal,self.soilMoistWet),self.humidityMedium),self.lightintensityNormal),irrigation_short)
        #Fuzzy Inference for Short Irrigation
        self.Rule_13 = np.fmin(np.fmin(self.tempFresh,self.soilMoistDry),irrigation_medium)
        self.Rule_14 = np.fmin(np.fmin(self.tempNormal,self.soilMoistHumide),irrigation_medium) 
        self.Rule_15 = np.fmin(np.fmin(self.tempHot,self.soilMoistWet),irrigation_medium)
        self.Rule_16 = np.fmin(np.fmin(self.humidityLow,self.lightintensityBright),irrigation_medium)  
        self.Rule_17 = np.fmin(np.fmin(self.humidityMedium,self.lightintensityNormal),irrigation_medium)
        self.Rule_18 = np.fmin(np.fmin(np.fmin(self.tempHot,self.humidityHigh),self.lightintensityBright),irrigation_medium)
        self.Rule_19 = np.fmin(np.fmin(np.fmin(np.fmin(self.tempFresh,self.soilMoistDry),self.humidityHigh),self.lightintensityDark),irrigation_medium)
        #Fuzzy Inference for Long Irrigation
        self.Rule_20 = np.fmin(np.fmin(self.tempNormal,self.soilMoistDry),irrigation_long)
        self.Rule_21 = np.fmin(np.fmin(self.tempHot,self.soilMoistDry),irrigation_long)
        self.Rule_22 = np.fmin(np.fmin(self.tempHot,self.soilMoistHumide),irrigation_long)
        self.Rule_23 = np.fmin(np.fmin(self.tempVeryHot,self.soilMoistDry),irrigation_long)
        self.Rule_24 = np.fmin(np.fmin(self.tempVeryHot,self.soilMoistHumide),irrigation_long)
        self.Rule_25 = np.fmin(np.fmin(self.tempVeryHot,self.soilMoistWet),irrigation_long)
        self.Rule_26 = np.fmin(np.fmin(self.humidityLow,self.lightintensityVeryBright),irrigation_long)
        self.Rule_27 = np.fmin(np.fmin(np.fmin(self.tempNormal,self.humidityLow),self.soilMoistDry),irrigation_long)
        self.Rule_28 = np.fmin(np.fmin(np.fmin(self.tempVeryHot,self.humidityLow),self.lightintensityVeryBright),irrigation_long)
        self.Rule_29 = np.fmin(np.fmin(np.fmin(np.fmin(self.tempHot,self.soilMoistHumide),self.humidityLow),self.lightintensityVeryBright),irrigation_long)

        self.out_short = np.fmax(np.fmax(np.fmax(np.fmax(np.fmax(np.fmax(np.fmax(np.fmax(np.fmax(np.fmax(np.fmax(self.Rule_1, self.Rule_2), self.Rule_3), self.Rule_4), self.Rule_5), self.Rule_6), self.Rule_7), self.Rule_8), self.Rule_9), self.Rule_10), self.Rule_11), self.Rule_12)
        self.out_medium = np.fmax(np.fmax(np.fmax(np.fmax(np.fmax(np.fmax(self.Rule_13, self.Rule_14), self.Rule_15), self.Rule_16), self.Rule_17), self.Rule_18), self.Rule_19)
        self.out_long = np.fmax(np.fmax(np.fmax(np.fmax(np.fmax(np.fmax(np.fmax(np.fmax(np.fmax(self.Rule_20, self.Rule_21), self.Rule_22), self.Rule_23), self.Rule_24), self.Rule_25), self.Rule_26), self.Rule_27), self.Rule_28), self.Rule_29)
    
    def inference_plot(self):
        self.irrigation = np.zeros_like(x_irrigation)
        fig, ax0 = plt.subplots(figsize = (7, 4))
        ax0.fill_between(x_irrigation, self.irrigation, self.out_short, label='short', facecolor = 'r', alpha = 0.7)
        ax0.plot(x_irrigation, irrigation_short, 'r', linestyle = '-')
        ax0.fill_between(x_irrigation, self.irrigation, self.out_medium, label='medium', facecolor = 'g', alpha = 0.7)
        ax0.plot(x_irrigation, irrigation_medium, 'g', linestyle = '-')
        ax0.fill_between(x_irrigation, self.irrigation, self.out_long, label='long', facecolor = 'b', alpha = 0.7)
        ax0.set_xlabel('Irrigation')
        ax0.plot(x_irrigation, irrigation_long, 'b', linestyle = '-')
        ax0.legend()
        ax0.set_title('Fuzzy Inference Output for Irrigation')
        plt.tight_layout()

    def defuzzification(self):
        self.out_irrigation = np.fmax(np.fmax(self.out_short, self.out_medium),self.out_long)
        self.defuzzified  = fuzz.defuzz(x_irrigation, self.out_irrigation, 'centroid')

        if (self.defuzzified > 0) & (self.defuzzified<= 17):
            print("Short Irrigation")
        elif (self.defuzzified > 17) & (self.defuzzified <= 37.5):
            print ("Medium Irrigation")
        elif (self.defuzzified > 37.5) & (self.defuzzified <= 60):
                print("Long Irrigation.")

        print(f'Irrigation = {self.defuzzified:.2f} mn')

    def defuzzification_plot(self):
        self.result = fuzz.interp_membership(x_irrigation, self.out_irrigation, self.defuzzified)

        fig, ax0 = plt.subplots(figsize=(7, 4))

        ax0.plot(x_irrigation, irrigation_short, 'r', linewidth = 0.5, linestyle = '--')
        ax0.plot(x_irrigation, irrigation_medium, 'g', linewidth = 0.5, linestyle = '--')
        ax0.plot(x_irrigation, irrigation_long, 'b', linewidth = 0.5, linestyle = '--')

        ax0.fill_between(x_irrigation, self.irrigation, self.out_irrigation, facecolor = 'Orange', alpha = 0.7)
        ax0.plot([self.defuzzified , self.defuzzified], [0, self.result], 'k', linewidth = 1.5, alpha = 0.9)
        ax0.set_title('Centroid Deffuzification')

        plt.tight_layout()
        #plt.show()
        
           
```

## Use-Case 1 : Warm, Humid Conditions with Low Light
-  **Scenario**: An overcast, warm day with high humidity. The soil is moderately moist.
- **Inputs**:
    -  Temperature: 25°C
    - Soil Moisture: 50%
    - Humidity: 70%
    - Light Intensity: 200 units
- **Expected Output**: Medium to low irrigation level as the evapotranspiration rate would be lower due to less light and moderate temperature.

```python
Green_House_Irrigation_1 = Fuzzy_Irrigation(CURRENT_TEMP = 25,CURRENT_SOIL_MOIST =50,CURRENT_HUMIDITY = 70 ,CURRENT_LIGHT_INTENSITY = 200)
Green_House_Irrigation_1.fuzzification()
```
